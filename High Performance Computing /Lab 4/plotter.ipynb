```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

# Increase resolutin of plots
plt.rcParams['figure.dpi'] = 200
```

### Cluster

```python
import os 
pwd = os.getcwd()
print(pwd)

path_Q1 = pwd + '/Q1/data.csv'
path_Q2 = pwd + '/Q2/data.csv'
path_Q3 = pwd + '/Q3/data.csv'


dataQ1 = pd.read_csv(path_Q1)
dataQ2 = pd.read_csv(path_Q2)
dataQ3 = pd.read_csv(path_Q3)

num_cores = [0, 2,4,8,12,16]

# open files at path_Q1, path_Q2, path_Q3 and remove all rows where log2(PROB_SIZE) > 28
dataQ1 = dataQ1[dataQ1['PROB_SIZE'] <= 2**28]
dataQ2 = dataQ2[dataQ2['PROB_SIZE'] <= 2**28]
dataQ3 = dataQ3[dataQ3['PROB_SIZE'] <= 2**28]

# Value of all entries in df['PROB_NAME'] in Q1 --> 'PI_TRAPEZOIDAL', Q2 --> 'SUM_VECTORS', Q3 --> 'QUICK_SORT'
dataQ1['PROB_NAME'] = 'PI_TRAPEZOIDAL'
dataQ2['PROB_NAME'] = 'SUM_VECTORS'
dataQ3['PROB_NAME'] = 'QUICK_SORT'

# export to csv files as data_updated.csv for all three questions as data_updatedQ1.csv, data_updatedQ2.csv, data_updatedQ3.csv
dataQ1.to_csv('data_updatedQ1.csv', index=False)
dataQ2.to_csv('data_updatedQ2.csv', index=False)
dataQ3.to_csv('data_updatedQ3.csv', index=False)
    

# Plot df['ALG_TIME'] vs log2(df['PROB_SIZE']) for all num_cores for Q1, Q2, Q3
#Q1
plt.figure(1)
for i in num_cores:
    df = dataQ1[dataQ1['NUMBER_CORES'] == i]
    label = 'NUM_CORES = ' + str(i)
    if i == 0:
        label = 'NUM_CORES = 1(SERIAL)'
    plt.plot(np.log2(df['PROB_SIZE']), df['ALG_TIME'],'-o', label = label,markersize=5)
plt.xlabel('$log_2$(PROB_SIZE)')
plt.ylabel('Algorithm Time')
plt.title('Algorithm Time vs PROB_SIZE for Q1A')
plt.legend()
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig('Q1_cluster.png')
plt.show()

#Q2
plt.figure(2)
for i in num_cores:
    df = dataQ2[dataQ2['NUMBER_CORES'] == i]
    label = 'NUM_CORES = ' + str(i)
    if i == 0:
        label = 'NUM_CORES = 1(SERIAL)'
    plt.plot(np.log2(df['PROB_SIZE']), df['ALG_TIME'],'-o', label = label ,markersize=5)
plt.xlabel('$log_2$(PROB_SIZE)')
plt.ylabel('Algorithm Time')
plt.title('Algorithm Time vs PROB_SIZE for Q2A')
plt.legend()
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig('Q2_cluster.png')
plt.show()


#Q3
plt.figure(3)
for i in num_cores:
    df = dataQ3[dataQ3['NUMBER_CORES'] == i]
    label = 'NUM_CORES = ' + str(i)
    if i == 0:
        label = 'NUM_CORES = 1(SERIAL)'
    plt.plot(np.log2(df['PROB_SIZE']), df['ALG_TIME'],'-o', label = label,markersize=5)
plt.xlabel('$log_2$(PROB_SIZE)')
plt.ylabel('Algorithm Time')
plt.title('Algorithm Time vs PROB_SIZE for Q3A')
plt.legend()
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig('Q3_cluster.png')
plt.show()

    

# Plot df['TOTAL_TIME'] vs log2(df['PROB_SIZE']) for all num_cores for Q1, Q2, Q3
#Q1
plt.figure(4)
for i in num_cores:
    df = dataQ1[dataQ1['NUMBER_CORES'] == i]
    label = 'NUM_CORES = ' + str(i)
    if i == 0:
        label = 'NUM_CORES = 1(SERIAL)'
    plt.plot(np.log2(df['PROB_SIZE']), df['TOTAL_TIME'],'-o', label = label,markersize=5)
plt.xlabel('$log_2$(PROB_SIZE)')
plt.ylabel('Total Time')
plt.title('Total Time vs PROB_SIZE for Q1A')
plt.legend()
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig('Q1t_cluster.png')
plt.show()

#Q2
plt.figure(5)
for i in num_cores:
    df = dataQ2[dataQ2['NUMBER_CORES'] == i]
    label = 'NUM_CORES = ' + str(i)
    if i == 0:
        label = 'NUM_CORES = 1(SERIAL)'
    plt.plot(np.log2(df['PROB_SIZE']), df['TOTAL_TIME'],'-o', label = label,markersize=5)
plt.xlabel('$log_2$(PROB_SIZE)')
plt.ylabel('Total Time')
plt.title('Total Time vs PROB_SIZE for Q2A')
plt.legend()
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig('Q2t_cluster.png')
plt.show()

#Q3
plt.figure(6)
for i in num_cores:
    df = dataQ3[dataQ3['NUMBER_CORES'] == i]
    label = 'NUM_CORES = ' + str(i)
    if i == 0:
        label = 'NUM_CORES = 1(SERIAL)'
    plt.plot(np.log2(df['PROB_SIZE']), df['TOTAL_TIME'],'-o', label = label,markersize=5)
plt.xlabel('$log_2$(PROB_SIZE)')
plt.ylabel('Total Time')
plt.title('Total Time vs PROB_SIZE for Q3A')
plt.legend()
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.savefig('Q3t_cluster.png')
plt.show()

        
```
