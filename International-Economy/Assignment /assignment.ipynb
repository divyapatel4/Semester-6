```python

df_gdp = df_gdp.drop(df_gdp.iloc[:, 4:44], axis = 1)
df_imports = df_imports.drop(df_imports.iloc[:, 4:44], axis = 1)
df_exports = df_exports.drop(df_exports.iloc[:, 4:44], axis = 1)

# drop last column from the dataframe 'Unnamed: 66'
df_gdp = df_gdp.drop(columns=['Unnamed: 66'])
df_imports = df_imports.drop(columns=['Unnamed: 66'])
df_exports = df_exports.drop(columns=['Unnamed: 66'])


# add one column developed country to the dataframe -- 1 for developed country and 0 for developing country check the list of countries
df_gdp['Developed'] = np.where(df_gdp['Country Name'].isin(developed_countries), 1, 0)

df_gdp.head(10)
```
