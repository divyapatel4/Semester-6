```python

df_gdp = df_gdp.drop(df_gdp.iloc[:, 4:44], axis = 1)
df_imports = df_imports.drop(df_imports.iloc[:, 4:44], axis = 1)
df_exports = df_exports.drop(df_exports.iloc[:, 4:44], axis = 1)

# drop last column from the dataframe 'Unnamed: 66'
df_gdp = df_gdp.drop(columns=['Unnamed: 66'])
df_imports = df_imports.drop(columns=['Unnamed: 66'])
df_exports = df_exports.drop(columns=['Unnamed: 66'])


# add one column developed country to the dataframe -- 1 for developed country and 0 for developing country check the list of countries
df_gdp['Developed'] = np.where(df_gdp['Country Name'].isin(developed_countries), 1, 0)

df_gdp.head(10)
```

```python
# sample 40 countries randomly from the dataset and make new df --> df_sample for gdp, exports and imports

countries = df_gdp['Country Name'].unique()

df_sample_gdp = df_gdp[df_gdp['Country Name'].isin(countries)]
df_sample_exports = df_exports[df_exports['Country Name'].isin(countries)]
df_sample_imports = df_imports[df_imports['Country Name'].isin(countries)]


for country in countries:
    if country not in df_sample_gdp['Country Name'].unique():
        df_sample_exports = df_sample_exports[df_sample_exports['Country Name'] != country]
        df_sample_imports = df_sample_imports[df_sample_imports['Country Name'] != country]
    if country not in df_sample_exports['Country Name'].unique():
        df_sample_gdp = df_sample_gdp[df_sample_gdp['Country Name'] != country]
        df_sample_imports = df_sample_imports[df_sample_imports['Country Name'] != country]
    if country not in df_sample_imports['Country Name'].unique():
        df_sample_gdp = df_sample_gdp[df_sample_gdp['Country Name'] != country]
        df_sample_exports = df_sample_exports[df_sample_exports['Country Name'] != country]
    
```

```python
df_sample_gdp = df_sample_gdp.dropna()
df_sample_gdp.shape
```

```python
df_sample_exports = df_sample_exports.dropna()
df_sample_exports.shape
```

```python
df_sample_imports = df_sample_imports.dropna()
df_sample_imports.shape
```

```python


gdp = df_sample_gdp
imports = df_sample_imports
exports = df_sample_exports

for country in countries:
    # if the country is in all the three dataframes then do nothing else remove the country from all the dataframes
    if country in gdp['Country Name'].unique() and country in imports['Country Name'].unique() and country in exports['Country Name'].unique():
        pass
    else:
        gdp = gdp[gdp['Country Name'] != country]
        imports = imports[imports['Country Name'] != country]
        exports = exports[exports['Country Name'] != country]




# sort all the dataframes by country code
gdp = gdp.sort_values(by=['Country Code'],ignore_index=True)
imports = imports.sort_values(by=['Country Code'],ignore_index=True)
exports = exports.sort_values(by=['Country Code'],ignore_index=True)

# reset index of all the dataframes

```

```python
trade_open = gdp.copy();   
# trade_openness = imports + exports / gdp

for year in range(2000,2022,1):
    trade_open[str(year)] = (imports[str(year)]+exports[str(year)])/gdp[str(year)]
```

```python
trade_open = trade_open.drop(['Indicator Name','Indicator Code','2000'],axis=1)
```

```python
# economic growth rate
# (curr-prev)*100/prev
eco_growth = gdp.copy()

for year in range(2001,2022,1):
    eco_growth[str(year)] = ((gdp[str(year)]-gdp[str(year-1)]))/gdp[str(year-1)]
    
```

```python
eco_growth = eco_growth.drop(['Indicator Name','Indicator Code','2000'],axis=1)
```

#### Spearman Correlation

```python
# spearman corealtion between trade openness and economic growth
import scipy.stats as stats
from scipy.stats import spearmanr, pearsonr

df_tradeopen_ecogrowth = pd.DataFrame(columns = ['Country Name', 'Developed', 'Spearman Correlation'])

for country in developed_countries:
    tradeopen = trade_open[trade_open['Country Name'] == country]
    trade_openval = []
    for i in range (2011,2021,1):
        trade_openval.append(tradeopen[str(i)].values[0])
    ecogrowth = eco_growth[eco_growth['Country Name'] == country]
    eco_growthval = []
    for i in range (2011,2021,1):
        eco_growthval.append(ecogrowth[str(i)].values[0])
    # spearman_corr, p_value = spearmanr(trade_openval, eco_growthval)
    spearman_corr = pd.DataFrame([trade_openval, eco_growthval]).T.corr().T.iloc[0, 1]
    
    df_tradeopen_ecogrowth = pd.concat([df_tradeopen_ecogrowth, pd.DataFrame({'Country Name': country, 'Developed': 1, 'Spearman Correlation': spearman_corr}, index=[0])], ignore_index=True)
    
for country in developing_countries:
    tradeopen = trade_open[trade_open['Country Name'] == country]
    ecogrowth = eco_growth[eco_growth['Country Name'] == country]
    trade_openval = [];eco_growthval = [];
    
    for i in range (2011,2021,1):
        trade_openval.append(tradeopen[str(i)].values[0])
        eco_growthval.append(ecogrowth[str(i)].values[0])
        
    spearman_corr,_ = spearmanr(trade_openval, eco_growthval)

    df_tradeopen_ecogrowth = pd.concat([df_tradeopen_ecogrowth, pd.DataFrame({'Country Name': country, 'Developed': 0, 'Spearman Correlation': spearman_corr}, index=[0])], ignore_index=True)



avg_corelation_developed = df_tradeopen_ecogrowth[df_tradeopen_ecogrowth['Developed'] == 1]['Spearman Correlation'].mean()
avg_corelation_developing = df_tradeopen_ecogrowth[df_tradeopen_ecogrowth['Developed'] == 0]['Spearman Correlation'].mean()
print('For developed countries the average corelation is: ', avg_corelation_developed)
print('For developing countries the average corelation is: ', avg_corelation_developing)
df_tradeopen_ecogrowth.head(50)

# trade_openval

```

#### Granger Causality Test

```python
# do gragner causality test for the developed countries and    developing countries
import statsmodels.tsa.stattools as ts
import statsmodels.api as sm
from statsmodels.tsa.stattools import grangercausalitytests


# do similar analysis for developed countries and developing countries seperately with chi square test and make df with chi square values and p values
df_tradeopen_ecogrwoth_gragner = pd.DataFrame(columns = ['Country Name', 'Developed', 'Gragner Causality(chi2 value)','p-value'])

for country in developed_countries:
    tradeopen = trade_open[trade_open['Country Name'] == country]
    trade_openval = []
    for i in range (2011,2021,1):
        trade_openval.append(tradeopen[str(i)].values[0])
        
    ecogrowth = eco_growth[eco_growth['Country Name'] == country]
    eco_growthval = []
    for i in range (2011,2021,1):
        eco_growthval.append(ecogrowth[str(i)].values[0])
        
    df = pd.DataFrame([trade_openval, eco_growthval]).T
    df.columns = ['trade_open', 'eco_growth']
    df.index = pd.date_range(start='2011', periods=10, freq='Y')
    # print(df)
    result = grangercausalitytests(df, maxlag=1, verbose=False)
    result = result[1][0]['ssr_ftest']
    chi_square = result[0]
    p_value = result[1]
    print(result)
    df_tradeopen_ecogrwoth_gragner = pd.concat([df_tradeopen_ecogrwoth_gragner, pd.DataFrame({'Country Name': country, 'Developed': 1, 'Gragner Causality(chi2 value)': chi_square, 'p-value': p_value}, index=[0])], ignore_index=True)
    
    

for country in developing_countries:
    tradeopen = trade_open[trade_open['Country Name'] == country]
    ecogrowth = eco_growth[eco_growth['Country Name'] == country]
    trade_openval = [];eco_growthval = [];
    
    for i in range (2011,2021,1):
        trade_openval.append(tradeopen[str(i)].values[0])
        eco_growthval.append(ecogrowth[str(i)].values[0])
        
    df = pd.DataFrame([trade_openval, eco_growthval]).T
    df.columns = ['trade_open', 'eco_growth']
    df.index = pd.date_range(start='2011', periods=10, freq='Y')
    # print(df)
    result = grangercausalitytests(df, maxlag=1, verbose=False)
    result = result[1][0]['ssr_ftest']
    chi_square = result[0]
    p_value = result[1]
    print(result)
    df_tradeopen_ecogrwoth_gragner = pd.concat([df_tradeopen_ecogrwoth_gragner, pd.DataFrame({'Country Name': country, 'Developed': 0, 'Gragner Causality(chi2 value)': chi_square, 'p-value': p_value}, index=[0])], ignore_index=True)

        
df_tradeopen_ecogrwoth_gragner.to_csv('tradeopen_ecogrowth_gragner.csv',index=False)
average_p_value_developed = df_tradeopen_ecogrwoth_gragner[df_tradeopen_ecogrwoth_gragner['Developed'] == 1]['p-value'].mean()
average_p_value_developing = df_tradeopen_ecogrwoth_gragner[df_tradeopen_ecogrwoth_gragner['Developed'] == 0]['p-value'].mean()

print('For developed countries the average p-value is: ', average_p_value_developed)
print('For developing countries the average p-value is: ', average_p_value_developing)
```

```python
# Test the hypothesis that mean trade openness of developed countries is greater than that of developing countries using z test(two sample test) at 5 per cent level of significance.

mean_trade_openness_developed = trade_open[trade_open['Country Name'].isin(developed_countries)]
mean_trade_openness_developing = trade_open[trade_open['Country Name'].isin(developing_countries)]

# add a new column as 'mean trade openness' which will contain the mean trade openness a country from all year from 2011 to 2020
for row in range(0,len(mean_trade_openness_developed)):
    sum = 0
    for year in range(2011,2021,1):
        sum += mean_trade_openness_developed.iloc[row][str(year)]
    mean_trade_openness_developed.loc[row,'mean trade openness'] = sum/10

for row in range(0,len(mean_trade_openness_developing)):
    sum = 0
    for year in range(2011,2021,1):
        sum += mean_trade_openness_developing.iloc[row][str(year)]
    mean_trade_openness_developing.loc[row,'mean trade openness'] = sum/10
    

# print mean of 'mean trade openness' for developed and developing countries
print('Mean trade openness of developed countries: ',mean_trade_openness_developed['mean trade openness'].mean())
print('Mean trade openness of developing countries: ',mean_trade_openness_developing['mean trade openness'].mean())
    


```

## B. Test the hypothesis that mean trade openness of developed countries is greater than that of developing countries using Z test (two sample test) at 5 per cent level of significance.

```python
# Now do the z test that is two sample test to check if the mean trade openness of developed countries is greater than that of developing countries
import math
# H0: mean trade openness of developed countries <= mean trade openness of developing countries
# H1: mean trade openness of developed countries > mean trade openness of developing countries
developed_countries_mean_tradeopeness = mean_trade_openness_developed['mean trade openness'].mean()
developing_countries_mean_tradeopeness = mean_trade_openness_developing['mean trade openness'].mean()
print('Mean trade openness of developed countries: ',developed_countries_mean_tradeopeness)
print('Mean trade openness of developing countries: ',developing_countries_mean_tradeopeness)

devlpd_mto = mean_trade_openness_developed['mean trade openness'].to_numpy()
devlpng_mto = mean_trade_openness_developing['mean trade openness'].to_numpy()

# drop na values
devlpd_mto = devlpd_mto[~np.isnan(devlpd_mto)]
devlpng_mto = devlpng_mto[~np.isnan(devlpng_mto)]

z_score , p_value = sm.stats.ztest(devlpd_mto, devlpng_mto, alternative='larger')
print('z-score: ', z_score)
print('p-value: ', p_value)

# z-score:  3.626869180264856
# p-value:  0.00014343927203929228
# therefore we reject the null hypothesis and conclude that mean trade openness of developed countries is greater than that of developing countries
```

## C. Test the hypothesis that the mean trade openness of developed countries is not equal to that of developing countries using test (two sample test) at 5 per cent level of significance.

```python
# Test the hypothesis that the mean trade openness of developed countries is not equal to that of developing countries using z test (two sample test) at 5 per cent level of significance.

# H0: mean trade openness of developed countries = mean trade openness of developing countries
# H1: mean trade openness of developed countries != mean trade openness of developing countries
 
z_score , p_value = sm.stats.ztest(devlpd_mto, devlpng_mto, alternative='two-sided')
print('z-score: ', z_score)
print('p-value: ', p_value)

# we get:
# z-score:  3.626869180264856
# p-value:  0.00028687854407858455
# which is less than 0.05 so we reject the null hypothesis and conclude that the mean trade openness of developed countries is not equal to that of developing countries
```
