```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def c_dot(F,V,C_in,C):
    a = F*C_in/V
    b = F/V
    return a-b*C

def C_int(C_in,C_o,F,V,t):
    return C_in - (C_in-C_o)*np.exp(-F/V*t)
```

```python
# plot c_t vs t
C_0 = 10
C_in = 3
F = 5*10**8
V = 10**12

# plot C for t = 0 to 10 with timestep 0.1
t = np.arange(0, 10000, 0.01)
C = C_int(C_in, C_0, F, V, t)


# find x where C = 5 with a minimum tolerance
x_point = 0
index = 0
for i in range(len(C)):
    if (abs(C[i]-5) < 0.00001):
        x_point = i
        break
print('C = 5 at t = ', t[x_point])


plt.plot(t, C, color = 'royalblue')
plt.xlabel('time (days)')
plt.ylabel('Concentration C(t)')
plt.title('Concentration vs Time')
plt.yticks(np.arange(0, 11, 1))
plt.xlim(0, 10000)
plt.ylim(0, 10)
# set light grey grid lines
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.show()
```

```python
def x_dot(k1,x):
    return -k1*x

def y_dot(k1,k2,x,y):
    return k1*x - k2*y

def x_func(k1,x_0,t):
    return x_0*np.exp(-k1*t)

def y_func(k1,k2,x_0,y_0,t):
    return (k1*x_0/(k1-k2))*(np.exp(-k2*t)-np.exp(-k1*t))
```

```python
x_0 = 1;y_0 = 0;k1 = 0.6931;k2 = 0.0231;

t = np.arange(0,100,0.01)

x = x_func(k1,x_0,t)
y = y_func(k1,k2,x_0,y_0,t)

# plot x and y vs t
plt.plot(t,x,label='Medicine Concentration in GI Tract (x(t))')
plt.plot(t,y,label='Medicine Concentration in Blood (y(t))')
plt.xlabel('time (hours)')
plt.ylabel('Medicine Concentration')
plt.title('Medicine Concentration vs Time')
plt.xlim(-1,100)
plt.ylim(-0.04,1)
plt.grid(color='lightgrey', linestyle='-', linewidth=0.5)
plt.legend()
plt.show()
    
```
