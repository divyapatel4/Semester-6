```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def rate_of_change(x):  
    r = 1; a=1; b=2;
    return x*(r-a*(x-b)**2)

# plot rate vs x 
x = np.linspace(-0.5, 3.25, 10000)
# make a horizontal line at y=0
plt.plot([x[0], x[-1]], [0, 0], 'k--')
plt.plot(x, rate_of_change(x))
plt.xlabel('x')
# plt.ylabel('rate of change')
# ylabel is x_dot in latex 
plt.ylabel('$\dot{x}$')
plt.show()
```

```python
def euler_integrate(func,x0,t):
    h = t[1]-t[0]
    x = np.zeros(len(t))
    x[0] = x0
    for i in range(len(t)-1):
        x[i+1] = x[i] + h*func(x[i])
    return x


r = 1; a=1; b=2;
point_x = b - np.sqrt(r/a) 
print(point_x)

# point_a = point_x - 0.2
# point_b = point_x + 0.2
# # these are the initial conditions x0 

# # Case A
# x0 = point_a
t = np.linspace(0, 5, 1000)
# x = euler_integrate(rate_of_change, x0, t)
# plt.plot(t, x, label='x0 = %.2f' % x0)

# # Case B
# x0 = point_b
# x = euler_integrate(rate_of_change, x0, t)
# plt.plot(t, x, label='x0 = %.2f' % x0)

# # Case C 
# x0 = point_x - 0.1
# x = euler_integrate(rate_of_change, x0, t)
# plt.plot(t, x, label='x0 = %.2f' % x0)

# # Case D
# x0 = point_x + 0.1
# x = euler_integrate(rate_of_change, x0, t)
# plt.plot(t, x, label='x0 = %.2f' % x0)

# Case E 
x0 = point_x + 0.01
x = euler_integrate(rate_of_change, x0, t)
plt.plot(t, x, label='$x_0$ = %.2f' % x0)

# Case F
x0 = point_x - 0.01
x = euler_integrate(rate_of_change, x0, t)
plt.plot(t, x, label='$x_0$ = %.2f' % x0)

# Case G
x0 = point_x + 0.001
x = euler_integrate(rate_of_change, x0, t)
plt.plot(t, x, label='$x_0$ = %.3f' % x0)

# Case H
x0 = point_x - 0.001
x = euler_integrate(rate_of_change, x0, t)
plt.plot(t, x, label='$x_0$ = %.3f' % x0)


plt.grid(color='lightgrey', linestyle='--', linewidth=0.5)
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()




```
