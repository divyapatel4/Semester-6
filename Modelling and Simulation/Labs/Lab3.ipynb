```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
def c_dot(F,V,C_in,C):
    a = F*C_in/V
    b = F/V
    return a-b*C

def C_int(C_in,C_o,F,V,t):
    return C_in - (C_in-C_o)*np.exp(-F/V*t)
```

```python
# plot c_t vs t 
C_0 = 10;C_in = 3;F = 5*10**8; V = 10**12

# plot C for t = 0 to 10 with timestep 0.1
t = np.arange(0,10000,0.01)
C = C_int(C_in,C_0,F,V,t)



# find x where C = 5 with a minimum tolerance 
x_point = 0; index = 0
for i in range(len(C)):
    if(abs(C[i]-5)<0.0001):
        x_point = i
        break
print('C = 5 at t = ',t[x_point])


plt.plot(t, C)
plt.xlabel('t')
plt.ylabel('C')
plt.title('C vs t')
plt.yticks(np.arange(0, 11, 1))
# palce a marker at x_point with a label
plt.plot(t[x_point],C[x_point],'ro')
plt.text(t[x_point],C[x_point],'C = 5 at t = '+str(t[x_point]))
    
plt.show()
```

```python
def time_taken(C_in,C,C_o,F,V):
    return np.log((C_in-C_o)/(C_in-C))*V/F
```

```python
C_0 = 10
C = 5
C_in = 3
F = 5*10**8
V = 10**12

time = time_taken(C_in,C,C_0,F,V)

print ('Time taken for C to reach 5 is ',time)
```

```python
def x_dot(k1,x):
    return -k1*x

def y_dot(k1,k2,x,y):
    return k1*x - k2*y

def x_func(k1,x_0,t):
    return x_0*np.exp(-k1*t)

def y_func(k1,k2,x_0,y_0,t):
    return (k1*x_0/(k1-k2))*(np.exp(-k2*t)-np.exp(-k1*t))
```

```python
x_0 = 1;y_0 = 0;k1 = 0.6931;k2 = 0.0231;

t = np.arange(0,100,0.01)

x = x_func(k1,x_0,t)
y = y_func(k1,k2,x_0,y_0,t)

# plot x and y vs t
plt.plot(t,x,label='x')
plt.plot(t,y,label='y')
plt.xlabel('t')
plt.ylabel('x and y')
plt.title('x and y vs t')
plt.legend()
plt.show()
    
```

```python
# when k1 = k2 
k2 = k1 + 0.0001

t = np.arange(0,10,0.01)

x = x_func(k1,x_0,t)
y = y_func(k1,k2,x_0,y_0,t)

# plot x and y vs t
plt.plot(t,x,label='x')
plt.plot(t,y,label='y')
plt.xlabel('t')
plt.ylabel('x and y')
plt.title('x and y vs t')
plt.legend()
plt.show()

```

```python
# for case of continuous supply of x 

k1 = 0.6931;k2 = 0.0231; I = 1.0; x_0 = 0; y_0 = 0;

def x_dot(k1,x,I):
    return -k1*x + I

def y_dot(k1,k2,x,y):
    return k1*x - k2*y

def x_func(k1,x_0,t,I):
    return I/k1*(1-np.exp(-k1*t))

def y_func(k1,k2,x_0,y_0,t,I):
    a = I/k2*(1-np.exp(-k2*t)) 
    b =I/(k2-k1)*(np.exp(-k1*t) - np.exp(-k2*t))
    return a-b

t = np.arange(0,5,0.001    )
x = x_func(k1,x_0,t,I)
y = y_func(k1,k2,x_0,y_0,t,I)

# plot x and y vs t
plt.plot(t,x,label='x')
plt.plot(t,y,label='y')
plt.xlabel('t')
plt.ylabel('x and y')
plt.title('x and y vs t')
plt.legend()
plt.show()


```
